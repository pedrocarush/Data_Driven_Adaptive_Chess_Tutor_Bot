```python
import pandas as pd
import numpy as np
import os
import csv
```

```python
BASE_FOLDER = '../data/3_lichess_csv_parsed'
INPUT_CSV = f'{BASE_FOLDER}/Elo_1200_1300.csv'
```

```python
df = pd.read_csv(INPUT_CSV, index_col=1, header=0)
```

```python
opening_dict = {
    
}
```

```python
def opening_stats(row: pd.Series,opening_dict:dict):
    
    if row['Opening'] not in opening_dict:
        opening_dict[row['Opening']] = {
            'Wins': 0,
            'Losses': 0,
            'Draws': 0
        }
    
    if row['Result'] == 1:
        opening_dict[row['Opening']]['Wins'] += 1
    elif row['Result'] == 0:
        opening_dict[row['Opening']]['Losses'] += 1
    else:
        opening_dict[row['Opening']]['Draws'] += 1
```

```python
df.apply(opening_stats, axis=1, args=(opening_dict,))
```

```python
opening_dict
```

```python
with open('Openings.csv', 'w') as csvfile: 
    writer = csv.DictWriter(csvfile, fieldnames = ['Opening', 'Wins', 'Losses', 'Draws']) 
    writer.writeheader() 
    for key in opening_dict.keys():
        writer.writerow({'Opening': key, 'Wins': opening_dict[key]['Wins'], 'Losses': opening_dict[key]['Losses'], 'Draws': opening_dict[key]['Draws']})
```
